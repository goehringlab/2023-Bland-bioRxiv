```python
import sys
sys.path.append('../../../src')
from helpers import *
%matplotlib notebook
```

```python
# Load reference profiles
cytbg = np.loadtxt('../../Raw analysis/Quantification model calibration/cytbg.txt')
membg1 = np.loadtxt('../../Raw analysis/Quantification model calibration/membg1.txt')
membg2 = np.loadtxt('../../Raw analysis/Quantification model calibration/membg2.txt')
membg3 = np.loadtxt('../../Raw analysis/Quantification model calibration/membg3.txt')
membg4 = np.loadtxt('../../Raw analysis/Quantification model calibration/membg4.txt')
membg5 = np.loadtxt('../../Raw analysis/Quantification model calibration/membg5.txt')
membg6 = np.loadtxt('../../Raw analysis/Quantification model calibration/membg6.txt')
```

### Membg

```python
def plot(ax, membg, c, label, linewidth=1):
    ax.plot(membg / max(membg), c=c, linewidth=linewidth, label=label)

fig, ax = plt.subplots()
plot(ax, membg1, 'tab:blue', 'PAR-2(wt)')
plot(ax, membg2, 'tab:orange', 'PAR-2(C56S)')
plot(ax, membg3, 'tab:green', 'PAR-2(L109R)')
plot(ax, membg4, 'tab:red', 'PAR-2(L50R)')
plot(ax, membg5, 'tab:purple', 'PAR-2(wt, single\ncopy)')
plot(ax, membg6, 'k', 'Full dataset')
     
ax.legend(fontsize=8, title='Training data')
ax.set_yticks([0, 1])
ax.tick_params(axis='both', labelsize=8)
ax.set_xlabel('Position')
ax.set_ylabel('Normalised intensity')
ax.set_title('Membrane signal profile', fontsize=10)
fig.set_size_inches(4, 3)
fig.tight_layout()
fig.savefig('Figs/membg_meany.png', dpi=600, transparent=True)
```
