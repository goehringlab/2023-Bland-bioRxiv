### Quantification model calibrations

- Displaying cytoplasmic and membrane signal profile calibrations
- Cytoplasmic calibration was performed using images of cytoplasmic mNG
- Several membrane calibration were performed using a range of PAR-2 images, which are compared below. Generally the shape is very similar for all the different calibrations
- The actual analysis is performed in _Run/calibration_cyt.ipynb_ and _Run/calibration_mem.ipynb_, were just loading the resulting txt files here

```python
import os

import matplotlib.pyplot as plt
import numpy as np

from src import nb_setup

nb_setup()
os.makedirs("Figs", exist_ok=True)

%matplotlib inline
```

```python
# Load reference profiles
cytbg = np.loadtxt("../../../data/calibration_cyt.txt")
membg1 = np.loadtxt("../../../data/calibration_mem1.txt")
membg2 = np.loadtxt("../../../data/calibration_mem2.txt")
membg3 = np.loadtxt("../../../data/calibration_mem3.txt")
membg4 = np.loadtxt("../../../data/calibration_mem4.txt")
membg5 = np.loadtxt("../../../data/calibration_mem5.txt")
membg6 = np.loadtxt("../../../data/calibration_mem6.txt")
```

### Membg

```python
def plot(ax, membg, c, label, linewidth=1):
    ax.plot(membg / max(membg), c=c, linewidth=linewidth, label=label)


fig, ax = plt.subplots()
plot(ax, membg1, "tab:blue", "PAR-2(wt)")
plot(ax, membg2, "tab:orange", "PAR-2(C56S)")
plot(ax, membg3, "tab:green", "PAR-2(L109R)")
plot(ax, membg4, "tab:red", "PAR-2(L50R)")
plot(ax, membg5, "tab:purple", "PAR-2(wt, single\ncopy)")
plot(ax, membg6, "k", "Full dataset")

# ax.legend(fontsize=8, title='Training data')
ax.set_yticks([0, 1])
ax.tick_params(axis="both", labelsize=8)
ax.set_xlabel("Position", labelpad=-5)
ax.set_ylabel("Normalized\nintensity", labelpad=-5)
ax.set_xticks([0, 50])
# ax.set_title('Membrane signal profile', fontsize=10)
fig.set_size_inches(2.5, 2.5)
fig.subplots_adjust(bottom=0.2, left=0.2)
fig.savefig("Figs/membg_comparison.png", dpi=600, transparent=True)
fig.savefig("Figs/membg_comparison.pdf", transparent=True)
```

```python
fig, ax = plt.subplots()
lw = 1
ax.plot([], color="k", label="Full\ndataset", linewidth=lw)
ax.plot([], color="tab:blue", label="WT", linewidth=lw)
ax.plot([], color="tab:orange", label="C56S", linewidth=lw)
ax.plot([], color="tab:green", label="L109R", linewidth=lw)
ax.plot([], color="tab:red", label="L50R", linewidth=lw)
ax.plot([], color="tab:purple", label="WT,\nsingle\ncopy", linewidth=lw)

legend = ax.legend(fontsize=7, title="Training data", loc="upper left", frameon=False)
plt.setp(legend.get_title(), fontsize="8")
legend.get_frame().set_edgecolor("k")
legend.get_frame().set_linewidth(0.5)
ax.axis("off")
fig.set_size_inches(2.5, 2.5)
fig.subplots_adjust(bottom=0.2, left=0.2)
fig.savefig("Figs/membg_comparison_legend.png", dpi=600, transparent=True)
fig.savefig("Figs/membg_comparison_legend.pdf", transparent=True)
```
